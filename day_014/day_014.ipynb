##### Explain the difference between map, filter, and reduce.
map function takes in two parameter: map(function, iterable). It takes a function and applies it on an iterable

filter function takes two parameters just like map but the outcome of the function parameter is a boolean(True or False) value thereby filtering the iterable. filter(function, iterable)

reduce function is imported from the functool library and takes in two parameters just like map and filter but return a single value. reduce(function, iterable)

```python
from functools import reduce

def call(func, iterable):
    result = func(iterable)
    return result

def cube(x):
    return x ** 3

def is_even(lst):
    lst_i = [i for i in lst if i % 2 == 0]
    return lst_i

def multiply(x, y):
    return x * y

lst = [1, 2, 3, 4, 5]

print("Example of map function:")
cubed_numbers = call(map, cube, lst)
print(list(cubed_numbers))

print("\nExample of filter function:")
filtered_numbers = call(filter, is_even, lst)
print(list(filtered_numbers))

print("\nExample of reduce function:")
product = call(reduce, multiply, lst)
print(product)
```

```python
# Use filter to filter out countries containing 'land'.
from countries import countries


def countries_with_land(country):
    for part in country:
        if 'land' in part.lower():  
            return False  
    return True


filtered_countries = list(filter(countries_with_land, countries))
print(filtered_countries)
```

```python
# Sort out the ten most spoken languages by location.
from collections import Counter
all_languages = [
    language for country in country_data for language in country["languages"]]

language_counts = Counter(all_languages)

top_languages = language_counts.most_common(10)

print("Top 10 Most Spoken Languages by Location:")
for language, count in top_languages:
    print(f"{language}: {count} countries")
```

```python
# Sort out the ten most populated countries.
sorted_countries = sorted(country_data, key=lambda x: x["population"], reverse=True)

top_populated_countries = sorted_countries[:10]


print("Top 10 Most Populated Countries:")
for country in top_populated_countries:
    print(f"{country['name']}: {country['population']} people")
```
